```python
#calculate class priors for 11 classes
priors=[]
features=12
labels=[i for i in range(0,11)]
classes=[]
for i in (labels):
    priors.append(sum(wine_white['quality']==i))
    classes.append(priors[i])
    priors[i]=round(priors[i]/N_wine,3)
priors
classes[6]
```

```python
#estimate mean vector of dim-11
mean_wine=np.zeros((features,len(labels)))
mean_wine.shape
for i in range(len(labels)):
    mean_wine[:,i]= round(np.mean(w[:,i]),3)
mean_wine[:,0]
#mean vector
```

```python
#covariance
v1=np.zeros(((features,features,len(labels))))
for i in range(len(labels)):
    v1[:,:,i]=np.cov(w,rowvar=False)
v1[:,:,0]
v1.shape
#covariance matrices
```
